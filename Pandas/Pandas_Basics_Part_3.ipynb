# Merge

```python
data1 = pd.DataFrame({"A":[1,2,3,4], "B":[50,60,70,80]})
data2 = pd.DataFrame({"C":[10,20,30,40], "B":[50,60,100,80]})

pd.merge(data2, data1, how="left")
```

```python
pd.merge(data2, data1, how="right")
```

```python
pd.merge(data2, data1, how="outer", indicator=True)
```

```python
data1 = pd.DataFrame({"A":[1,2,3,4], "B":[50,60,70,80]})
data2 = pd.DataFrame({"A":[10,20,30,40], "B":[50,60,100,80]})

pd.merge(data2, data1, left_index=True, right_index=True)
```

```python
pd.merge(data2, data1, left_index=True, right_index=True, suffixes=("past", "future"))
```

# Concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([10,20,30,40])

pd.concat([sr1, sr2])
```

```python
data1 = pd.DataFrame({"A":[1,2,3,4], "B":[50,60,70,80]})
data2 = pd.DataFrame({"C":[10,20,30,40], "B":[50,60,100,80]})

pd.concat([data1, data2])
```

```python
data1 = pd.DataFrame({"A":[1,2,3,4], "B":[50,60,70,80]})
data2 = pd.DataFrame({"C":[10,20,30,40], "D":[50,60,100,80]})

pd.concat([data1, data2], axis=1)
```

```python
data1 = pd.DataFrame({"A":[1,2,3,4], "B":[50,60,70,80]})
data2 = pd.DataFrame({"C":[10,20], "D":[50,60]})

pd.concat([data1, data2], axis=1, join="inner")
```

```python
pd.concat([data1, data2], axis=1, join="outer")
```

```python
data1 = pd.DataFrame({"A":[1,2,3,4], "B":[50,60,70,80]})
data2 = pd.DataFrame({"C":[10,20], "D":[50,60]})

pd.concat([data1, data2], axis=0, keys=["one", "two"])
# pd.concat([data1, data2], axis=1, keys=["one", "two"])
```

```python
data1 = pd.DataFrame({"A":[1,2,3,4], "B":[50,60,70,80]})
data2 = pd.DataFrame({"C":[10,20]})

pd.concat([data1, data2])
```

# GroupBy

```python
data = pd.DataFrame({"name": ["A", "B", "C", "D", "D", "A", "C", "B", "D", "C", "D"],
                     "s1":[21,24,87,45,98,23,76,12,78,34,67],
                    "s2":[43,65,17,23,12,35,54,34,76,45,87]})
data
```

```python
new_data = data.groupby("name")
new_data
```

```python
for x,y in new_data:
    print(x)
    print(y, '\n---------------')
```

```python
new_data.get_group("D")
```

# Join

```python
df1 = pd.DataFrame({'C':[21,22], 'D':[101,102]})
df2 = pd.DataFrame({'A':[1,2,3,4,5], 'B':[6,7,8,9,10]})

df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1 = pd.DataFrame({'C':[21,22], 'D':[101,102]},index=["A","B"])
df2 = pd.DataFrame({'A':[1,2,3,4,5], 'B':[6,7,8,9,10]})

df1.join(df2)
```

```python
df1 = pd.DataFrame({'C':[21,22], 'D':[101,102]},index=["A","B"])
df2 = pd.DataFrame({'A':[1,2,3,4,5], 'B':[6,7,8,9,10]},index=[100,200,300,"B",500])

# df1.join(df2, how=('left'))
# df1.join(df2, how=('right'))
# df1.join(df2, how=('outer'))
df1.join(df2, how=('inner'))
```

```python
df1 = pd.DataFrame({'A':[1,2,3,4,5], 'B':[6,7,8,9,10]},index=["A","B","C","D","E"])
df2 = pd.DataFrame({'C':[21,22], 'D':[101,102]})

df1.join(df2)
```

```python
df1 = pd.DataFrame({'A':[1,2,3,4,5], 'B':[6,7,8,9,10]},index=["A","B","C","D","E"])
df2 = pd.DataFrame({'C':[21,22], 'D':[101,102]},index=['B','D'])

df1.join(df2)
```

```python
df1 = pd.DataFrame({'C':[21,22], 'D':[101,102]},index=["A","B"])
df2 = pd.DataFrame({'A':[1,2,3,4,5], 'D':[6,7,8,9,10]},index=[100,200,300,"B",500])

df1.join(df2, how=('outer'), rsuffix=17, lsuffix=1700)
```

# melt()

```python
df1 = pd.DataFrame({"Day":['Mon','Tue','Wed','Thurs','Fri'],"Math":[100,98,99,97,100],"English":[85,86,82,91,83],"Science":[90,92,91,94,94]})
df1
```

```python
pd.melt(df1)
```

```python
pd.melt(df1, id_vars=['Day'], var_name=('Subject'), value_name=('Score'))
```

# pivot()

```python
df1 = pd.DataFrame({"Day":['Mon','Tue','Wed','Thurs','Fri'],
                    "Student":['A','B','C','D','E'],"Math":[100,98,99,97,100],
                    "English":[85,86,82,91,83],"Science":[90,92,91,94,94]})

df1.pivot(index='Day',columns='Student')
```

```python
df1.pivot(index='Day',columns='Student',values=['English'])
```

```python
df1 = pd.DataFrame({"Day":['Mon','Tue','Mon','Mon','Tue'],
                    "Student":['A','B','A','B','B'],"Math":[100,98,99,97,100],
                    "English":[85,86,82,91,83],"Science":[90,92,91,94,94]})

df1.pivot_table(index='Student',columns='Day',aggfunc='mean')
```

```python
df1.pivot_table(index='Student',columns='Day',aggfunc='median',margins=True)
```
